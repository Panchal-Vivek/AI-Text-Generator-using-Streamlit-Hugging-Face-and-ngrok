```python
!pip install -q streamlit pyngrok requests

!ngrok config add-authtoken "2voyjzPBEuJ3PM2FDMRyKeNXqVm_6yBT5KPFQhwkXiK6QzL2K"

with open("app.py", "w") as f:
    f.write('''
import streamlit as st
import requests

API_TOKEN = "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"

MODEL_OPTIONS = {
    "Mistral 7B Instruct": "mistralai/Mistral-7B-Instruct-v0.3"
}

st.set_page_config(page_title="AI Text Generator", page_icon="✨")
st.title("🤖 AI Text Generator (Hugging Face)")
st.markdown("Generate high-quality text using lightweight models under 10GB.")

prompt = st.text_area("💬 Enter your prompt:")

selected_model = st.selectbox("🔍 Choose a model:", list(MODEL_OPTIONS.keys()))

max_tokens = st.slider("📝 Max New Tokens", 50, 1000, value=150)
temperature = st.slider("🔥 Temperature (Creativity)", 0.1, 1.0, value=0.7)
top_p = st.slider("🎲 Top-p (Diversity)", 0.1, 1.0, value=0.9)

def generate_text(prompt, model_id, max_tokens, temperature, top_p):
    API_URL = f"https://api-inference.huggingface.co/models/mistralai/Mixtral-8x7B-Instruct-v0.1"
    headers = {"Authorization": f"Bearer {API_TOKEN}"}
    payload = {
        "inputs": prompt,
        "parameters": {
            "max_new_tokens": max_tokens,
            "temperature": temperature,
            "top_p": top_p,
            "do_sample": True,
            "return_full_text": False
        }
    }
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

if st.button("🚀 Generate"):
    if not prompt.strip():
        st.warning("Please enter a prompt.")
    else:
        with st.spinner("Generating response..."):
            model_id = MODEL_OPTIONS[selected_model]
            result = generate_text(prompt, model_id, max_tokens, temperature, top_p)

            if isinstance(result, list):
                st.success("✅ Response:")
                st.write(result[0]["generated_text"])
            elif "error" in result:
                st.error(f"❌ Error: {result['error']}")
            else:
                st.error("❌ Unexpected error occurred.")

if st.button("🧹 Clear"):
    st.experimental_rerun()
''')

# 🚀 Launch Streamlit with ngrok
from pyngrok import ngrok
import time
import threading

def run_app():
    !streamlit run app.py &

threading.Thread(target=run_app).start()
time.sleep(5)

public_url = ngrok.connect(8501)
print(f"🌍 Open your Streamlit app: {public_url}")
```
